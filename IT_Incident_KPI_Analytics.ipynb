```python
mask = pd.to_datetime(df['opened_at'], errors = 'coerce').isna() & df['opened_at'].notna()
df.loc[mask, 'opened_at'].unique()
```

```python
mask = pd.to_datetime(df['closed_at'], errors = 'coerce').isna() & df['closed_at'].notna()
df.loc[mask, 'closed_at'].unique()
```

```python
df['opened_at'] = pd.to_datetime(df['opened_at'], errors = 'coerce', dayfirst = True)
df['closed_at'] = pd.to_datetime(df['closed_at'], errors = 'coerce', dayfirst = True)
```

```python
df[['opened_at', 'closed_at']].dtypes
```

```python
df['number'].nunique(), len(df)
```

```python
inc_df = df[['number']].drop_duplicates().reset_index(drop = True)
```

```python
inc_df.head()
```

```python
inc_df['number'].info()
```

```python
inc_df['opened_at'] = (df.groupby('number')['opened_at'].min().values)
```

```python
inc_df['closed_at'] = (df.groupby('number')['closed_at'].max().values)
```

```python
inc_df.head()
```

```python
resolved_series = df[df['incident_state'] == 'Resolved'].groupby('number')['resolved_at'].min()
```

```python
inc_df['resolved_at'] = inc_df['number'].map(resolved_series)
```

```python
inc_df.head(7)
```

```python
made_sla_series = (df[df['incident_state'] == 'Closed'].groupby('number')['made_sla'].first())
inc_df['made_sla'] = inc_df['number'].map(made_sla_series)
```

```python
inc_df.head()
```

```python
inc_df['reopen_count'] = (df.groupby('number')['reopen_count'].max().values)
```

```python
inc_df.head()
```

```python
print(inc_df['reopen_count'].unique())
print(df['reopen_count'].unique())
#An INC may temporarily have reopen_count = 5,
#but later it gets updated to 6 or 8,
#so 5 (and 7) never appear in the final incident-level table.
```

```python
print(inc_df['reopen_count'].nunique())
print(df['reopen_count'].nunique())
```

```python
inc_df['resolution_time_hours_KPI01'] = ((inc_df['resolved_at'] - inc_df['opened_at']).dt.total_seconds()/3600)
```

```python
inc_df.head()
```

```python
inc_df['closure_time_hours_KPI02'] = ((inc_df['closed_at'] - inc_df['opened_at']).dt.total_seconds()/3600)
inc_df.head()
```

```python
inc_df['resolution_time_hours_KPI01'] = inc_df['resolution_time_hours_KPI01'].round(2)
inc_df['closure_time_hours_KPI02'] = inc_df['closure_time_hours_KPI02'].round(2)
inc_df.head()
```

```python
#inc_df['made_sla'].dtype
inc_df['SLA_compliance_%_KPI03'] = (((inc_df['made_sla'] == True).sum())/(inc_df['number'].count())) * 100
inc_df.head()
```

```python
#inc_df['sla_compliance_pct_KPI03_test'] = (
#    (inc_df['made_sla'] == True).sum() / inc_df['number'].count()
#) * 100
#inc_df.head()
```

```python
inc_df['SLA_compliance_%_KPI03'] = inc_df['SLA_compliance_%_KPI03'] .round(2)
inc_df.drop(columns = ['sla_compliance_pct_KPI03_test'], inplace =  True)
inc_df.head()
```

```python
inc_df['Reopen_rate_KPI04'] = (((inc_df['reopen_count']>0).sum())/ inc_df['number'].count()) * 100
inc_df['Reopen_rate_KPI04'] = inc_df['Reopen_rate_KPI04'].round(2)
```

```python
inc_df.head()
```

```python
inc_df['Reopen_rate_KPI04'].unique()
```

```python
reopen_rate_KPI04 = round(
    (inc_df['reopen_count'] > 0).sum() / inc_df['number'].count() * 100,
    2
)
```

```python
inc_df.drop(columns=['Reopen_rate_KPI04'], inplace=True)
```

```python
reopen_rate_KPI04
```

```python
sla_compliance_KPI03 = round((inc_df['made_sla'] == True).sum()/inc_df['number'].count() * 100, 2)
```

```python
sla_compliance_KPI03
```

```python
inc_df.drop(columns = ['SLA_compliance_%_KPI03'], inplace = True)
```

<h3>Trend Graphs</h3>

```python
import matplotlib.pyplot as plt

inc_vol_monthly = (inc_df.set_index('opened_at').resample('ME').size())
#.size() means: “Count how many rows exist per month”

plt.figure(figsize = (9,6))
plt.plot(inc_vol_monthly.index, inc_vol_monthly.values, marker = 'o')
for x, y in zip(inc_vol_monthly.index, inc_vol_monthly.values):
    plt.text(x, y+120, str(y), fontsize=8)
plt.xlabel('Months')
plt.ylabel('Incident volume')
plt.title('Monthly Incident volume trend')
plt.show()
```

```python
avg_resol_time_monthly = (inc_df.set_index('opened_at')['resolution_time_hours_KPI01'].resample('ME').mean().round(2))
avg_resol_time_monthly
plt.figure(figsize = (9,6))
plt.plot(avg_resol_time_monthly.index, avg_resol_time_monthly.values, marker = 'o')
for x, y in zip(avg_resol_time_monthly.index, avg_resol_time_monthly.values):
    plt.text(x, y+20, str(y), fontsize=8)
plt.xlabel('Months')
plt.ylabel('Average Resolution Time (in Hours)')
plt.title('Monthly Average Incident Resolution trend')
plt.show()
```

```python
sla_compliance_monthly = (inc_df.set_index('opened_at')['made_sla'].resample('ME').mean()*100).round(2)

plt.figure(figsize = (9,6))
plt.plot(sla_compliance_monthly.index, sla_compliance_monthly.values, marker = 'o')
for x, y in zip(sla_compliance_monthly.index, sla_compliance_monthly.values):
    plt.text(x, y+1, str(y), fontsize=8)
plt.xlabel('Months')
plt.ylabel('SLA Compliance (%)')
plt.title('Monthly SLA Compliance trend')
plt.show()
```

```python
reopen_rate_monthly = (inc_df.set_index('opened_at')['reopen_count'].gt(0).resample('ME').mean()*100).round(2)

plt.figure(figsize = (9,6))
plt.plot(reopen_rate_monthly.index, reopen_rate_monthly.values, marker = 'o')
for x, y in zip(reopen_rate_monthly.index, reopen_rate_monthly.values):
    plt.text(x, y+0.1, str(y), fontsize=8)
plt.xlabel('Months')
plt.ylabel('Reopen rate (%)')
plt.title('Monthly Reopen Rate Trend')
plt.show()
```
